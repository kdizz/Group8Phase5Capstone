By leveraging machine learning (XGBoost), the project seeks to uncover key economic drivers influencing currency fluctuations and provide forward-looking insights. Kenya’s exchange rate has experienced significant fluctuations over the years, affecting trade, investment, and economic stability. ### Converting the columns to numeric 

```python
#We shall make the columns numeric except for Date which we shall convert to datetime and make an index prior to beginnning visualization.

for df in [dfaverage, dfclosing, dfmonthly, dfopening]:
    for col in df.columns:
        if col != 'Date':
            df[col] = pd.to_numeric(df[col], errors='coerce')
```

### Handling missing values



```python
#We shall replace the missing data by mean, as the deviations are and should be very low across all the dataframes
for df in [dfaverage, dfclosing, dfmonthly, dfopening]:
    for col in df.columns:
        if df[col].dtype in ['float64', 'int64']:
            if df[col].isnull().any():
                df[col].fillna(df[col].mean(), inplace=True)
```

```python
#confirming it worked.
dfmonthly.isna().sum()
```

### Converting the Date Column to datetime and setting it as an index

```python
#We shall convert the Date columns in all the datasets to allow for visualization.
for df in [dfaverage, dfclosing, dfmonthly, dfopening]:
    if 'Date' in df.columns:
        df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
        df.set_index('Date', inplace=True)
```

```python
import pandas as pd
import numpy as np
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import seaborn as sns
import matplotlib.pyplot as plt


# Load data
df = pd.read_csv("dfmonthly_modelling.csv")

# Clean column names (strip spaces)
df.columns = df.columns.str.strip()

# Drop 'Date'
df_model = df.drop(columns=["Date"])

# Set target and features
X_all = df_model.drop(columns=["exchange_rate"])
y = df_model["exchange_rate"]

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X_all, y, test_size=0.2, random_state=42)

# Train XGBoost
model_all = XGBRegressor(random_state=42)
model_all.fit(X_train, y_train)
y_pred_all = model_all.predict(X_test)

# Evaluate
print("Full Model:")
print("R²:", r2_score(y_test, y_pred_all))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_all)))
print("MAE:", mean_absolute_error(y_test, y_pred_all))

# Additional metrics
mape = np.mean(np.abs((y_test - y_pred_all) / y_test)) * 100
mse = mean_squared_error(y_test, y_pred_all)

print("MAPE:", mape)
print("MSE:", mse)


```

FUTURE FORCASTING USING XGBOOST

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Load and prepare data
df = pd.read_csv("dfmonthly_modelling.csv")
df['Date'] = pd.to_datetime(df['Date'])
df.sort_values('Date', inplace=True)
df.set_index('Date', inplace=True)

# Use only the exchange_rate column for univariate forecasting
data = df[['exchange_rate']].copy()

# Create lag features (e.g., past 12 months)
for lag in range(1, 13):
    data[f'lag_{lag}'] = data['exchange_rate'].shift(lag)

# Drop rows with NaN (due to lagging)
data.dropna(inplace=True)

# Train on all historical data before forecasting
X_train = data.drop(columns=['exchange_rate'])
y_train = data['exchange_rate']

# Train XGBoost
model = XGBRegressor(random_state=42)
model.fit(X_train, y_train)

# Start recursive forecasting from the last known point
last_known = data.iloc[-1:].copy()
future_predictions = []
future_dates = pd.date_range(start=data.index[-1] + pd.DateOffset(months=1), periods=24, freq='MS')

for date in future_dates:
    # Create new feature row using last known lags
    new_row = {}
    for lag in range(1, 13):
        new_row[f'lag_{lag}'] = last_known[f'lag_{lag}'].values[0] if lag > 1 else last_known['exchange_rate'].values[0]
    
    X_new = pd.DataFrame([new_row])
    y_pred = model.predict(X_new)[0]
    future_predictions.append(y_pred)

    # Update last_known with shifted lags
    new_lags = [y_pred] + list(last_known.iloc[0, :-1].values)  # shift forward
    new_entry = pd.DataFrame([new_lags], columns=['exchange_rate'] + [f'lag_{i}' for i in range(1, 13)])
    last_known = new_entry.copy()

# Create future DataFrame
future_df = pd.DataFrame({'Date': future_dates, 'Forecasted_Exchange_Rate': future_predictions})
future_df.set_index('Date', inplace=True)


# Plot historical + forecast
plt.figure(figsize=(14, 6))
plt.plot(data.index, data['exchange_rate'], label='Historical')
plt.plot(future_df.index, future_df['Forecasted_Exchange_Rate'], label='Forecast (2024–2025)', linestyle='--')
plt.title("XGBoost Exchange Rate Forecast: 2024–2025")
plt.xlabel("Date")
plt.ylabel("Exchange Rate")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Optionally save to CSV
future_df.to_csv("xgboost_forecast_2024_2025.csv")
```

This plot is a time series visualization showing how the exchange rate evolved historically and how it is expected to behave in the future based on your XGBoost model's forecast.

Key takeaways

• The forecasted exchange rate curve smoothly extends from the historical trend, indicating that the model generalizes well without overfitting. 
• The forecast plot for 2024–2025 suggests a continuation of recent trends, assuming macroeconomic conditions remain stable.
