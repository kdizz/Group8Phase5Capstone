# BUSINESS UNDERSTANDING

Exchange rates play a critical role in shaping the economic landscape of nations, influencing trade, investment, and overall economic stability. This project adopts an exploratory approach to examine how various factors impact the movements of Kenya’s exchange rate, specifically focusing on the USD/KES pair. The primary objective is to conduct a comprehensive exploratory data analysis (EDA) of the USD/KES exchange rate from 2003 to 2023, with the aim of identifying significant trends, seasonal patterns, and decomposing the time series to understand underlying structures. Additionally, the project seeks to develop and compare time series forecasting models to predict exchange rate movements for the year 2024 and validate these predictions against actual data from the same year. Another key objective is to apply supervised learning techniques to detect distinct volatility or trend regimes within the USD/KES exchange rate data, enabling a more nuanced understanding of its behavior over time. Finally, the study aims to identify and analyze macroeconomic indicators that have the most significant influence on exchange rate fluctuations, offering insights into the key drivers of currency dynamics in the Kenyan context.

# OBJECTIVES

1. To conduct exploratory data analysis (EDA) on the USD/KES exchange rate from 2003 to 2023, identifying major trends, seasonal patterns, and conducting time series decomposition.

2. To develop and compare time series forecasting models to predict exchange rate movements in 2024 and validate forecasts against actual 2024 data.

3. To apply supervised learning techniques to detect distinct volatility or trend regimes in USD/KES exchange rate data.

4. To determine the macroeconomic indicators with the greatest influence on exchange rate behavior.

### Converting the columns to numeric

```python
#We shall make the columns numeric except for Date which we shall convert to datetime and make an index prior to beginnning visualization.

for df in [dfaverage, dfclosing, dfmonthly, dfopening]:
    for col in df.columns:
        if col != 'Date':
            df[col] = pd.to_numeric(df[col], errors='coerce')
```

### Handling missing values



```python
#We shall replace the missing data by mean, as the deviations are and should be very low across all the dataframes
for df in [dfaverage, dfclosing, dfmonthly, dfopening]:
    for col in df.columns:
        if df[col].dtype in ['float64', 'int64']:
            if df[col].isnull().any():
                df[col].fillna(df[col].mean(), inplace=True)
```

```python
#confirming it worked.
dfmonthly.isna().sum()
```

### Converting the Date Column to datetime and setting it as an index

```python
#We shall convert the Date columns in all the datasets to allow for visualization.
for df in [dfaverage, dfclosing, dfmonthly, dfopening]:
    if 'Date' in df.columns:
        df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
        df.set_index('Date', inplace=True)
```

There seems to be a pattern in the movement and as such it would be best to overlay both plots to see the movements much clearer.

```python
#overlaying the bank interest rates and the exchage rate
sns.set_style("whitegrid")

fig, ax1 = plt.subplots(figsize=(14, 6))

sns.lineplot(data=dfmonthly, x="Date", y="Savings", ax=ax1, label="Savings", color="blue", linewidth=1)
sns.lineplot(data=dfmonthly, x="Date", y="Deposit", ax=ax1, label="Deposit", color="black", linewidth=1)
sns.lineplot(data=dfmonthly, x="Date", y="Lending ", ax=ax1, label="Lending", color="green", linewidth=1)
sns.lineplot(data=dfmonthly, x="Date", y="Overdraft ", ax=ax1, label="Overdraft", color="red", linewidth=1)

ax1.set_ylabel("Interest Rate", fontsize=12)
ax1.set_xlabel("Year", fontsize=12)
ax1.set_title("Bank Specific Interest Rates and Exchange Rate Over Time", fontsize=14, fontweight="bold")

ax2 = ax1.twinx()
sns.lineplot(data=dfmonthly, x="Date", y="exchange_rate", ax=ax2,
             label="Exchange Rate", color="purple", linewidth=2, linestyle="--", alpha=0.8)

ax2.set_ylabel("Exchange Rate", fontsize=12)

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()

ax1.legend(lines1 + lines2, labels1 + labels2, title="Indicator", loc="upper center", bbox_to_anchor=(0.5, -0.15), ncol=3)

ax1.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.6)
plt.tight_layout()
plt.subplots_adjust(bottom=0.2)

plt.show()
```

Key Takeaways;

1. The deposit interast rate tends to follow closely the exchange rate, we can attribute this to the fact that money is weakening, less people are investing and as such due to the low demand for fixed deposits, banks are forced to increase the interest rate to attract more money.

2. Lending and Overdraft have very little effect on the exchange rate.

```python
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
from sklearn.metrics import mean_squared_error, mean_absolute_error

# We will use the monthly data for the LSTM model
data = dfmonthly[['exchange_rate']].values

# Scaled the data to be between 0 and 1
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)

# Created the training and testing data sets
training_data_len = int(np.ceil( len(scaled_data) * .80 ))

train_data = scaled_data[0:training_data_len, :]
look_back = 60
test_data = scaled_data[training_data_len - look_back: , :]

def create_dataset(dataset, look_back=1):
    X, Y = [], []
    for i in range(len(dataset) - look_back - 1):
        a = dataset[i:(i + look_back), 0]
        X.append(a)
        Y.append(dataset[i + look_back, 0])
    return np.array(X), np.array(Y)

X_train, y_train = create_dataset(train_data, look_back)
X_test, y_test = create_dataset(test_data, look_back)

# Reshaped the data for the LSTM model
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

# Building the LSTM model

model = Sequential()
model.add(LSTM(128, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(Dropout(0.2))
model.add(LSTM(64, return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(25))
model.add(Dense(1))

# Compiled the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Trained the model
model.fit(X_train, y_train, batch_size=1, epochs=10)

# made predictions
predictions = model.predict(X_test)
predictions = scaler.inverse_transform(predictions)

y_test_actual = scaler.inverse_transform(y_test.reshape(-1, 1))
```

```python
# Evaluated the model
from sklearn.metrics import mean_squared_error, mean_absolute_error
rmse = np.sqrt(mean_squared_error(y_test_actual, predictions))
mae = mean_absolute_error(y_test_actual, predictions)
mse = mean_squared_error(y_test_actual, predictions)
print(f'RMSE: {rmse}')
print(f'MAE: {mae}')
```

- MAE = 5.65 on average the model's predictions are off by Ksh. 5.65 from the actual exchange rate price
- RMSE = 7.52 shows that model misses the true rate by 7.52.T his being higher than MAE suggests that some large prediction errors might be inflating the RMSE

```python
# a forecast for exchange rates

def predict_future(model, last_sequence, num_future_steps, scaler):
    future_predictions = []
    current_sequence = last_sequence.copy()

    for _ in range(num_future_steps):
        # Reshaped for prediction
        predicted_value = model.predict(current_sequence.reshape(1, look_back, 1))

        # Inverse transformed the prediction
        predicted_value_scaled = scaler.inverse_transform(predicted_value)
        future_predictions.append(predicted_value_scaled[0, 0])

        # removed the oldest value, added the new prediction
        current_sequence = np.append(current_sequence[:, 1:, :], predicted_value.reshape(1, 1, 1), axis=1)

    return future_predictions

# Number of future steps to predict
num_future_months = 24

# Got the last 'look_back' data points from the scaled data
last_sequence = scaled_data[-look_back:].reshape(1, look_back, 1)

# Predicted the future
future_forecast = predict_future(model, last_sequence, num_future_months, scaler)

# Created a date range for the future predictions
last_date = dfmonthly.index[-1]
future_dates = pd.date_range(start=last_date, periods=num_future_months + 1, freq='M')[1:]

# Created a DataFrame for the future forecast
future_forecast_df = pd.DataFrame({'exchange_rate': future_forecast}, index=future_dates)

```

```python
# modelled with variables that have a high correlation to exchange rates.

features = ['Deposit', '91-Day Tbill', '182-days Tbill', '364-days Tbill', 'exchange_rate']
data_multi = dfmonthly[features].values

# Scaled the data
scaler_multi = MinMaxScaler(feature_range=(0, 1))
scaled_data_multi = scaler_multi.fit_transform(data_multi)

# Created the training and testing data sets
training_data_len_multi = int(np.ceil( len(scaled_data_multi) * .80 ))

train_data_multi = scaled_data_multi[0:training_data_len_multi, :]
test_data_multi = scaled_data_multi[training_data_len_multi - look_back: , :] # Use the same look_back as before

# Created datasets with multiple input features
def create_dataset_multi(dataset, look_back=1):
    X, Y = [], []
    # Predicted the last column (exchange_rate)
    target_column_index = dataset.shape[1] - 1
    for i in range(len(dataset) - look_back - 1):

        a = dataset[i:(i + look_back), :]
        X.append(a)
        # targeted the exchange rate at the next step
        Y.append(dataset[i + look_back, target_column_index])
    return np.array(X), np.array(Y)

X_train_multi, y_train_multi = create_dataset_multi(train_data_multi, look_back)
X_test_multi, y_test_multi = create_dataset_multi(test_data_multi, look_back)

# Reshaped the data for the LSTM model
X_train_multi = np.reshape(X_train_multi, (X_train_multi.shape[0], X_train_multi.shape[1], X_train_multi.shape[2]))
X_test_multi = np.reshape(X_test_multi, (X_test_multi.shape[0], X_test_multi.shape[1], X_test_multi.shape[2]))


# Built the LSTM model with multiple input features
model_multi = Sequential()
model_multi.add(LSTM(128, return_sequences=True, input_shape=(X_train_multi.shape[1], X_train_multi.shape[2])))
model_multi.add(Dropout(0.2))
model_multi.add(LSTM(64, return_sequences=False))
model_multi.add(Dropout(0.2))
model_multi.add(Dense(25))
model_multi.add(Dense(1))

# Compiled the model
model_multi.compile(optimizer='adam', loss='mean_squared_error')

# Trained the model
print("\nTraining LSTM model with multiple correlated features...")
model_multi.fit(X_train_multi, y_train_multi, batch_size=1, epochs=10)

# Made the predictions
predictions_multi = model_multi.predict(X_test_multi)

# Inverse transformed the predictions and created a dummy array to inverse transform correctly.
dummy_array = np.zeros((predictions_multi.shape[0], len(features)))
dummy_array[:, features.index('exchange_rate')] = predictions_multi[:, 0]
predictions_multi_actual = scaler_multi.inverse_transform(dummy_array)[:, features.index('exchange_rate')]


# Inverse transformed the actual test values created a similar dummy array for the y_test_multi which is a single column
dummy_array_y_test = np.zeros((y_test_multi.shape[0], len(features)))
dummy_array_y_test[:, features.index('exchange_rate')] = y_test_multi
y_test_actual_multi = scaler_multi.inverse_transform(dummy_array_y_test)[:, features.index('exchange_rate')]

# Evaluated the model
rmse_multi = np.sqrt(mean_squared_error(y_test_actual_multi, predictions_multi_actual))
mae_multi = mean_absolute_error(y_test_actual_multi, predictions_multi_actual)

print(f'\nRMSE (Multi-feature model): {rmse_multi}')
print(f'MAE (Multi-feature model): {mae_multi}')

```
