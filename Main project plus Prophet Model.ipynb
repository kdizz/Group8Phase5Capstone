# BUSINESS UNDERSTANDING

Exchange rates play a critical role in shaping the economic landscape of nations, influencing trade, investment, and overall economic stability. This project adopts an exploratory approach to examine how various factors impact the movements of Kenyaâ€™s exchange rate, specifically focusing on the USD/KES pair. The primary objective is to conduct a comprehensive exploratory data analysis (EDA) of the USD/KES exchange rate from 2003 to 2023, with the aim of identifying significant trends, seasonal patterns, and decomposing the time series to understand underlying structures. Additionally, the project seeks to develop and compare time series forecasting models to predict exchange rate movements for the year 2024 and validate these predictions against actual data from the same year. Another key objective is to apply supervised learning techniques to detect distinct volatility or trend regimes within the USD/KES exchange rate data, enabling a more nuanced understanding of its behavior over time. Finally, the study aims to identify and analyze macroeconomic indicators that have the most significant influence on exchange rate fluctuations, offering insights into the key drivers of currency dynamics in the Kenyan context.

# OBJECTIVES

1. To conduct exploratory data analysis (EDA) on the USD/KES exchange rate from 2003 to 2023, identifying major trends, seasonal patterns, and conducting time series decomposition.

2. To develop and compare time series forecasting models to predict exchange rate movements in 2024 and validate forecasts against actual 2024 data.

3. To apply supervised learning techniques to detect distinct volatility or trend regimes in USD/KES exchange rate data.

4. To determine the macroeconomic indicators with the greatest influence on exchange rate behavior.

### Converting the columns to numeric

```python
#We shall make the columns numeric except for Date which we shall convert to datetime and make an index prior to beginnning visualization.

for df in [dfaverage, dfclosing, dfmonthly, dfopening]:
    for col in df.columns:
        if col != 'Date':
            df[col] = pd.to_numeric(df[col], errors='coerce')
```

### Handling missing values



```python
#We shall replace the missing data by mean, as the deviations are and should be very low across all the dataframes
for df in [dfaverage, dfclosing, dfmonthly, dfopening]:
    for col in df.columns:
        if df[col].dtype in ['float64', 'int64']:
            if df[col].isnull().any():
                df[col].fillna(df[col].mean(), inplace=True)
```

```python
#confirming it worked.
dfmonthly.isna().sum()
```

### Converting the Date Column to datetime and setting it as an index

```python
#We shall convert the Date columns in all the datasets to allow for visualization.
for df in [dfaverage, dfclosing, dfmonthly, dfopening]:
    if 'Date' in df.columns:
        df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
        df.set_index('Date', inplace=True)
```

```python
# Prophet requires the dataframe to have columns named 'ds' for the date and 'y' for the target variable
df_prophet = dfmonthly.reset_index().rename(columns={'Date': 'ds', 'exchange_rate': 'y'})

# Initialize and fit the Prophet model with all other columns as extra regressors
regressor_cols = [col for col in df_prophet.columns if col not in ['ds', 'y']]

model = Prophet()

# Add each regressor to the model
for col in regressor_cols:
    model.add_regressor(col)

# Fit the model
model.fit(df_prophet)

# Create a future dataframe for prediction
num_future_months = 12
future = model.make_future_dataframe(periods=num_future_months, freq='MS')

# We need to add the regressor values for the future dates.
for col in regressor_cols:
  future[col] = None
  future.loc[future['ds'].isin(df_prophet['ds']), col] = df_prophet.set_index('ds').loc[future.loc[future['ds'].isin(df_prophet['ds']), 'ds'], col].values
  future[col].fillna(df_prophet[col].iloc[-1], inplace=True)

# Make predictions
forecast = model.predict(future)

# Display the first few rows of the forecast
print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail())

# Plot the forecast
fig1 = model.plot(forecast)
plt.title("Prophet Forecast of Exchange Rate with Regressors")
plt.xlabel("Date")
plt.ylabel("Exchange Rate")
plt.show()

# Plot the components of the forecast
fig2 = model.plot_components(forecast)
plt.show()
```

The Prophet forecast for the exchange rate shows a consistent upward trend, particularly in recent years, with the model accurately reflecting historical data and predicting continued growth. The component breakdown reveals a strong long-term trend, distinct weekly and yearly seasonal patterns, and the additive influence of external regressors, all contributing to a comprehensive understanding of the exchange rate's behavior.

```python
# Prophet requires the dataframe to have columns named 'ds' for the date and 'y' for the target variable
df_prophet_selected = dfmonthly.reset_index().rename(columns={'Date': 'ds', 'exchange_rate': 'y'})

# Identify columns to exclude
exclude_cols = [
    'Imports',
    'Exports',
    'Foreign direct investment',
    'IBRD loans and IDA credits (DOD, current US$)',
    'Unemployment Rate'
]

# Identify regressor columns, excluding the ones specified
regressor_cols = [col for col in df_prophet_selected.columns if col not in ['ds', 'y'] + exclude_cols]

# Initialize and fit the Prophet model with selected columns as extra regressors
model_selected = Prophet()

# Add each selected regressor to the model
for col in regressor_cols:
    if col in df_prophet_selected.columns:
        model_selected.add_regressor(col)
    else:
        print(f"Warning: Regressor column '{col}' not found in the DataFrame.")


# Fit the model
model_selected.fit(df_prophet_selected)

# Create a future dataframe for prediction
num_future_months = 12
future = model_selected.make_future_dataframe(periods=num_future_months, freq='MS')

# We need to add the regressor values for the future dates.
for col in regressor_cols:
  if col in df_prophet_selected.columns:
    future[col] = None
    matching_dates = future.loc[future['ds'].isin(df_prophet_selected['ds']), 'ds']
    future.loc[future['ds'].isin(df_prophet_selected['ds']), col] = df_prophet_selected.set_index('ds').loc[matching_dates, col].values
    future[col].fillna(df_prophet_selected[col].iloc[-1], inplace=True)


# Make predictions
forecast_selected = model_selected.predict(future)

# Display the first few rows of the forecast
print(forecast_selected[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail())

# Plot the forecast
fig1 = model_selected.plot(forecast_selected)
plt.title("Prophet Forecast of Exchange Rate with Selected Regressors")
plt.xlabel("Date")
plt.ylabel("Exchange Rate")
plt.show()

# Plot the components of the forecast
fig2 = model_selected.plot_components(forecast_selected)
plt.show()
```

The Prophet forecast for the exchange rate shows a strong upward trend, particularly accelerating after 2018, with the model accurately capturing historical patterns and projecting continued growth. The breakdown into components reveals a consistent long-term increase, distinct weekly and yearly cyclical patterns, and a fluctuating, but generally diminishing, influence from additional external regressors.

```python
# Calculate the correlation matrix with the target variable
corr_with_target = dfmonthly.corr()['exchange_rate'].sort_values(ascending=False)

# Defining a threshold for high correlation
correlation_threshold = 0.8

# Identify columns with high absolute correlation
highly_correlated_cols = corr_with_target[abs(corr_with_target) > correlation_threshold].index.tolist()
highly_correlated_cols.remove('exchange_rate')

print("Highly correlated columns with 'exchange_rate':")
print(highly_correlated_cols)

# Preparing the data for Prophet, ensuring 'ds' and 'y' columns exist
df_prophet_correlated = dfmonthly.reset_index().rename(columns={'Date': 'ds', 'exchange_rate': 'y'})

# Initialize the Prophet model
model_correlated = Prophet()

# Add each highly correlated column as an extra regressor
for col in highly_correlated_cols:
    if col in df_prophet_correlated.columns:
        model_correlated.add_regressor(col)
    else:
        print(f"Warning: Highly correlated column '{col}' not found in the DataFrame.")


# Fit the model
model_correlated.fit(df_prophet_correlated)

# Creating a future dataframe for prediction
num_future_months = 12
future_correlated = model_correlated.make_future_dataframe(periods=num_future_months, freq='MS')

# Adding the regressor values for the future dates.
for col in highly_correlated_cols:
  if col in df_prophet_correlated.columns:
    future_correlated[col] = None
    df_prophet_correlated_indexed = df_prophet_correlated.set_index('ds')
    future_correlated.loc[future_correlated['ds'].isin(df_prophet_correlated['ds']), col] = df_prophet_correlated_indexed.loc[future_correlated.loc[future_correlated['ds'].isin(df_prophet_correlated['ds']), 'ds'], col].values
    future_correlated[col].fillna(df_prophet_correlated[col].iloc[-1], inplace=True)

# Make predictions
forecast_correlated = model_correlated.predict(future_correlated)

# Display the first few rows of the forecast
print(forecast_correlated[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail())

# Plot the forecast
fig1_correlated = model_correlated.plot(forecast_correlated)
plt.title("Prophet Forecast of Exchange Rate with Highly Correlated Regressors")
plt.xlabel("Date")
plt.ylabel("Exchange Rate")
plt.show()

# Plot the components of the forecast
fig2_correlated = model_correlated.plot_components(forecast_correlated)
plt.show()
```

The Prophet forecast for the exchange rate shows an increasing trend, especially since 2018, with the model projecting this upward movement to continue. The forecast components reveal a clear long-term upward trend, distinct weekly and yearly seasonal patterns, and a decreasing or negative influence from additional regressors over time. This decomposition helps in understanding the various factors driving the exchange rate's behavior.
