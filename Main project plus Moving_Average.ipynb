# BUSINESS UNDERSTANDING

Exchange rates play a critical role in shaping the economic landscape of nations, influencing trade, investment, and overall economic stability. This project adopts an exploratory approach to examine how various factors impact the movements of Kenyaâ€™s exchange rate, specifically focusing on the USD/KES pair. The primary objective is to conduct a comprehensive exploratory data analysis (EDA) of the USD/KES exchange rate from 2003 to 2023, with the aim of identifying significant trends, seasonal patterns, and decomposing the time series to understand underlying structures. Additionally, the project seeks to develop and compare time series forecasting models to predict exchange rate movements for the year 2024 and validate these predictions against actual data from the same year. Another key objective is to apply supervised learning techniques to detect distinct volatility or trend regimes within the USD/KES exchange rate data, enabling a more nuanced understanding of its behavior over time. Finally, the study aims to identify and analyze macroeconomic indicators that have the most significant influence on exchange rate fluctuations, offering insights into the key drivers of currency dynamics in the Kenyan context.

# OBJECTIVES

1. To conduct exploratory data analysis (EDA) on the USD/KES exchange rate from 2003 to 2023, identifying major trends, seasonal patterns, and conducting time series decomposition.

2. To develop and compare time series forecasting models to predict exchange rate movements in 2024 and validate forecasts against actual 2024 data.

3. To apply supervised learning techniques to detect distinct volatility or trend regimes in USD/KES exchange rate data.

4. To determine the macroeconomic indicators with the greatest influence on exchange rate behavior.

### Converting the columns to numeric

```python
#We shall make the columns numeric except for Date which we shall convert to datetime and make an index prior to beginnning visualization.

for df in [dfaverage, dfclosing, dfmonthly, dfopening]:
    for col in df.columns:
        if col != 'Date':
            df[col] = pd.to_numeric(df[col], errors='coerce')
```

### Handling missing values



```python
#We shall replace the missing data by mean, as the deviations are and should be very low across all the dataframes
for df in [dfaverage, dfclosing, dfmonthly, dfopening]:
    for col in df.columns:
        if df[col].dtype in ['float64', 'int64']:
            if df[col].isnull().any():
                df[col].fillna(df[col].mean(), inplace=True)
```

```python
#confirming it worked.
dfmonthly.isna().sum()
```

### Converting the Date Column to datetime and setting it as an index

```python
#We shall convert the Date columns in all the datasets to allow for visualization.
for df in [dfaverage, dfclosing, dfmonthly, dfopening]:
    if 'Date' in df.columns:
        df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
        df.set_index('Date', inplace=True)
```

```python
ts = dfmonthly['exchange_rate']
#Train-test split
train = ts[:-12]
test = ts[-12:]
```

```python
#Stationarity check using ADF test
adf_result = adfuller(train.dropna())
print("ADF Statistic:", adf_result[0])
print("p-value:", adf_result[1])

if adf_result[1] > 0.05:
    print("Training series is non-stationary. Applying first-order differencing.")
    train_model = train.diff().dropna()
    differenced = True
else:
    print("Training series is stationary. Proceeding without differencing.")
    train_model = train
    differenced = False

#Fitting MA model
ma_model = ARIMA(train_model, order=(0, 0, 1))
ma_result = ma_model.fit()
```

```python
#Forecast for the test period
diff_forecast = ma_result.get_forecast(steps=12).predicted_mean
if differenced:
    last_train_value = train.iloc[-1]
    forecast_original = [last_train_value + diff_forecast.iloc[0]]
    for i in range(1, len(diff_forecast)):
        forecast_original.append(forecast_original[i-1] + diff_forecast.iloc[i])
    forecast_original = pd.Series(forecast_original, index=test.index)
else:
    forecast_original = pd.Series(diff_forecast.values, index=test.index)

#Evaluation Metrics
mse = mean_squared_error(test, forecast_original)
rmse = np.sqrt(mse)
mae = mean_absolute_error(test, forecast_original)

print("\nEvaluation Metrics:")
print(f"MAE:  {mae:.4f}")
print(f"MSE:  {mse:.4f}")
print(f"RMSE: {rmse:.4f}")
```

The Mean Absolute Error (MAE) of 14.37 indicates that, on average, our predictions differ from the actual exchange rates by about 14.37 units. The Mean Squared Error (MSE), which is 286.68, penalizes larger errors more heavily by squaring the differences between predicted and actual values, making it useful for highlighting large deviations. The Root Mean Squared Error (RMSE), at 16.93, is simply the square root of the MSE and provides an error measure in the same units as the original data, making it more interpretable

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.stattools import adfuller
from sklearn.metrics import mean_squared_error, mean_absolute_error
import warnings
warnings.filterwarnings("ignore")

y = dfmonthly['exchange_rate']

exog_vars = ['Exports ', 'Imports ', 'Total Debt', 'Total Remittances ',
             'IBRD loans and IDA credits (DOD, current US$)',
             'Unemployment Rate', 'Deposit']
X = dfmonthly[exog_vars]
#Splitting into train and test
train_size = len(dfmonthly) - 12
y_train, y_test = y.iloc[:train_size], y.iloc[train_size:]
X_train, X_test = X.iloc[:train_size], X.iloc[train_size:]
#Checking stationarity
adf_result = adfuller(y_train.dropna())
p_value = adf_result[1]
d = 1 if p_value > 0.05 else 0
print(f"ADF p-value: {p_value:.4f} => Differencing: {d}")
#Performing grid search
best_score, best_cfg = float("inf"), None
for p in range(0, 3):
    for q in range(0, 3):
        try:
            model = SARIMAX(y_train, exog=X_train, order=(p, d, q),
                            enforce_stationarity=False, enforce_invertibility=False)
            results = model.fit(disp=False)
            forecast = results.get_forecast(steps=12, exog=X_test)
            forecast_mean = forecast.predicted_mean
            rmse = np.sqrt(mean_squared_error(y_test, forecast_mean))
            if rmse < best_score:
                best_score, best_cfg = rmse, (p, d, q)
            print(f"Tested SARIMAX{(p,d,q)} RMSE={rmse:.4f}")
        except:
            continue
#Best model display
print(f"\n Best SARIMAX Order: {best_cfg} with RMSE={best_score:.4f}")
#Training best model
best_p, best_d, best_q = best_cfg
model = SARIMAX(y_train, exog=X_train, order=(best_p, best_d, best_q),
                enforce_stationarity=False, enforce_invertibility=False)
results = model.fit()
#Forecasting on test
forecast = results.get_forecast(steps=12, exog=X_test)
forecast_mean = forecast.predicted_mean

mae = mean_absolute_error(y_test, forecast_mean)
mse = mean_squared_error(y_test, forecast_mean)
rmse = np.sqrt(mse)

print("\nðŸ“Š Evaluation Metrics:")
print(f"MAE:  {mae:.4f}")
print(f"MSE:  {mse:.4f}")
print(f"RMSE: {rmse:.4f}")
#Plotting
plt.figure(figsize=(12, 6))
plt.plot(y_train, label='Training Data')
plt.plot(y_test, label='Actual Test Data')
plt.plot(forecast_mean, label='Forecasted Data', color='red')
plt.title('Actual vs Forecasted Exchange Rate (Best SARIMAX)')
plt.xlabel('Date')
plt.ylabel('Exchange Rate')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

future_exog = X.iloc[-12:]
#refitting model
final_model = SARIMAX(y, exog=X, order=(best_p, best_d, best_q),
                      enforce_stationarity=False, enforce_invertibility=False).fit()
future_forecast = final_model.get_forecast(steps=12, exog=future_exog)
future_index = pd.date_range(start=y.index[-1] + pd.offsets.MonthEnd(1), periods=12, freq='M')
future_series = pd.Series(future_forecast.predicted_mean.values, index=future_index)
#Plotting
plt.figure(figsize=(12, 6))
plt.plot(y, label='Historical Exchange Rate')
plt.plot(future_series, label='12-Month Forecast', color='green')
plt.title('12-Month Exchange Rate Forecast (Future Projection)')
plt.xlabel('Date')
plt.ylabel('Exchange Rate')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
#Forecasted values
print("\nðŸ“… 12-Month Forecast:")
print(future_series)
```
