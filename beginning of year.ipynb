


DATA CLEANING


```python
#checking missing values
df.isnull().sum()
```

```python
# Impute missing values in 'Total Debt' using the median of the column
df['Total Debt'] = df['Total Debt'].fillna(df['Total Debt'].median())
```

```python
df.isnull().sum()
```

```python
#fixing the data type
#Ensure the Date column is a proper datetime type
df['Date'] = pd.to_datetime(df['Date'])
df.dtypes
```

```python
# Clean column names by stripping whitespace
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
df.head()
```





EXPLORATORY DATA  ANALYSIS



```python
# Line plot: Exchange Rate over time
plt.figure(figsize=(14, 6))
plt.plot(df['date'], df['exchange_rate'], marker='o')
plt.title('Exchange Rate Over Time')
plt.xlabel('Year')
plt.ylabel('Exchange Rate (KES/USD)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

This line graph above reveals a steady upward trend, indicating that the exchange rate has increased, meaning KES has depreciated over the years.
The line plot helps visualize long-term economic pressure on Kenya's currency


```python
#Plot Central Bank Rate and Exchange Rate
plt.figure(figsize=(14, 7))
plt.plot(df['date'], df['central_bank_rate'], marker='o', label='Central Bank Rate')
plt.plot(df['date'], df['exchange_rate'], marker='s', label='Exchange Rate')

# Plot formatting
plt.title("Central Bank Rate and Exchange Rate Over Time")
plt.xlabel("Year")
plt.ylabel("Rate")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

This plot helps assess whether interest rate policy is aligned with exchange rate trends.
It is Useful for economic forecasting, central bank strategy review, or investor analysis

```python
# Create the figure and axis
fig, ax1 = plt.subplots(figsize=(14, 7))

# Plot Exports and Imports on the primary Y-axis
ax1.plot(df['date'], df['exports'], color='green', marker='o', label='Exports')
ax1.plot(df['date'], df['imports'], color='red', marker='s', label='Imports')
ax1.set_xlabel("Year")
ax1.set_ylabel("Value in Ksh (Left Axis)", color='black')
ax1.tick_params(axis='y', labelcolor='black')
ax1.legend(loc='upper left')

# Create a secondary Y-axis for Exchange Rate
ax2 = ax1.twinx()
ax2.plot(df['date'], df['exchange_rate'], color='blue', marker='^', label='Exchange Rate')
ax2.set_ylabel("Exchange Rate (Right Axis)", color='blue')
ax2.tick_params(axis='y', labelcolor='blue')
ax2.legend(loc='upper right')

# Add title and layout settings
plt.title("Exports and Imports Over Time with Exchange Rate")
plt.grid(True)
plt.tight_layout()
plt.show()
```

This plot compares Exports, Imports, and the Exchange Rate over time using two vertical axes.
It helps assess the relationship between currency value and trade performance, which is critical for:
Policy analysis
Economic planning
Foreign exchange forecasting

```python
#  Trend Comparison Line Plot  Multiple Indicators
plt.figure(figsize=(14, 7))
plt.plot(df['date'], df['exchange_rate'], label='Exchange Rate')
plt.plot(df['date'], df['12-month_inflation(%)'], label='Inflation (%)')
plt.plot(df['date'], df['central_bank_rate'], label='Central Bank Rate')
plt.title("Trend Comparison of Key Economic Indicators Over Time")
plt.xlabel("Year")
plt.ylabel("Rate / Value")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

The plot allows one to visually compare changes over time across all three indicators.

it can also spot whether trends are moving together or diverging at key points.

```python
plt.figure(figsize=(14, 10))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix of Economic Indicators")
plt.tight_layout()
plt.show()
```

This heatmap visualizes the correlations between key economic indicators using the color gradients to visualize this:n

Blue cells represent strong positive relationships, where two variables increase together.
Red cells indicate strong negative correlations, where one variable increases as the other decreases.
White or light-colored cells show weak or no correlation
